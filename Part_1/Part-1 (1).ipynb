2. Convert the Dataframe into a temporary view called iot.

```python
rawDF.createOrReplaceTempView("iot_devices")
```

3. Count how many devices are there from each country and display the output.

```python
spark.sql("SELECT cn , COUNT(cn) FROM iot_devices GROUP BY cn").show(5, False)
```

4. Display all the countries whose carbon dioxide level is more than 1400. Sort the output in descending order.

```python
spark.sql("SELECT cn, c02_level FROM iot_devices WHERE c02_level > 1400 ORDER BY c02_level DESC").show(20, False)
```

5. Select all countries' devices with high-levels of C02 and group by cca3 and order by device_ids (Hint: For high CO2 level, the LCD status will be RED).

```python
spark.sql("SELECT cca3, count(*) as device_ids FROM iot_devices WHERE c02_level > 1400 GROUP BY cca3 ORDER BY device_ids").show(5, False)

# spark.sql("select cca3,count(*) as device_id from iot_devices where lcd == 'red' GROUP BY cca3  ORDER BY device_id").show()
```

```python
spark.sql("SELECT COUNT(device_id),cca3 from iot_devices where lcd == 'red' group by cca3 ORDER BY COUNT(device_id) DESC").show()
```
